### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

precios_tienda1 = tienda['Precio'].tolist()
precios_tienda2 = tienda2['Precio'].tolist()
precios_tienda3 = tienda3['Precio'].tolist()
precios_tienda4 = tienda4['Precio'].tolist()
costo_envios1 = tienda['Costo de envío'].tolist()
costo_envios2 = tienda2['Costo de envío'].tolist()
costo_envios3 = tienda3['Costo de envío'].tolist()
costo_envios4 = tienda4['Costo de envío'].tolist()

Ingreso_tienda1 = sum(precios_tienda1)
Ingreso_tienda2 = sum(precios_tienda2)
Ingreso_tienda3 = sum(precios_tienda3)
Ingreso_tienda4 = sum(precios_tienda4)
ventas_tienda1 = len(precios_tienda1)
ventas_tienda2 = len(precios_tienda2)
ventas_tienda3 = len(precios_tienda3)
ventas_tienda4 = len(precios_tienda4)

print(f'el ingreso total de la tienda 1: ${Ingreso_tienda1:,.2f}')
print(f'el ingreso total de la tienda 2: ${Ingreso_tienda2:,.2f}')
print(f'el ingreso total de la tienda 3: ${Ingreso_tienda3:,.2f}')
print(f'el ingreso total de la tienda 4: ${Ingreso_tienda4:,.2f}')

if ventas_tienda1 < ventas_tienda2 and ventas_tienda1 < ventas_tienda3 and ventas_tienda1 < ventas_tienda4:
  print(f'la tienda con menor ventas es la tienda 1 con {ventas_tienda1:,.0f} ventas')
elif ventas_tienda2 < ventas_tienda1 and ventas_tienda2 < ventas_tienda3 and ventas_tienda2 < ventas_tienda4:
  print(f'la tienda con menor ventas es la tienda 2 con {ventas_tienda2:,.0f} ventas')
elif ventas_tienda3 < ventas_tienda1 and ventas_tienda3 < ventas_tienda2 and ventas_tienda3 < ventas_tienda4:
  print(f'la tienda con menor ventas es la tienda 3 con {ventas_tienda3:,.0f} ventas')
else:
  print(f'la tienda con menor ventas es la tienda 4 con {ventas_tienda4:,.0f} ventas')

if Ingreso_tienda1 < Ingreso_tienda2 and Ingreso_tienda1 < Ingreso_tienda3 and Ingreso_tienda1 < Ingreso_tienda4:
  print(f'la tienda con menor ingreso es la tienda 1 con ${Ingreso_tienda1:,.2f}')
elif Ingreso_tienda2 < Ingreso_tienda1 and Ingreso_tienda2 < Ingreso_tienda3 and Ingreso_tienda2 < Ingreso_tienda4:
  print(f'la tienda con menor ingreso es la tienda 2 con ${Ingreso_tienda2:,.2f}')
elif Ingreso_tienda3 < Ingreso_tienda1 and Ingreso_tienda3 < Ingreso_tienda2 and Ingreso_tienda3 < Ingreso_tienda4:
  print(f'la tienda con menor ingreso es la tienda 3 con ${Ingreso_tienda3:,.2f}')
else:
  print(f'la tienda con menor ingreso es la tienda 4 con ${Ingreso_tienda4:,.2f}')
```

```python
import matplotlib.pyplot as plt

tiendas = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
ingresos = [Ingreso_tienda1, Ingreso_tienda2, Ingreso_tienda3, Ingreso_tienda4]
media_ingresos = sum(ingresos)/len(ingresos)

plt.figure(figsize=(10,6))
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
columns = plt.bar(tiendas, ingresos, color = 'skyblue', label = 'Ingresos por tienda')
plt.axhline(media_ingresos,color ='red', linestyle ='--', linewidth = 2, label = f'Media: ${media_ingresos:,.2f}')

plt.title('Ingresos totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylim(0, max(ingresos) * 1.15)

for i, ingreso in enumerate(ingresos):
    plt.text(i, ingreso + media_ingresos * 0.02, f"${ingreso:,.0f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

tienda['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True)

ingresos_por_fecha = tienda.groupby('Fecha de Compra')['Precio'].sum()
ingresos_suavizados = ingresos_por_fecha.rolling(window=7).mean()

x = (ingresos_por_fecha.index - ingresos_por_fecha.index[0]).days  # días desde la primera fecha
y = ingresos_por_fecha.values

slope, intercept, r_value, p_value, std_err = linregress(x, y)
tendencia = intercept + slope * x

plt.figure(figsize=(12, 6))
plt.plot(ingresos_por_fecha.index, ingresos_por_fecha.values, label='Original', alpha=0.4, linestyle='--')
plt.plot(ingresos_suavizados.index, ingresos_suavizados.values, color='blue', label='Media móvil (7 días)', linewidth=2)
plt.plot(ingresos_por_fecha.index, tendencia, color='red', linewidth=2,
         label=f'Tendencia lineal (R² = {r_value**2:.2f})')
plt.title('Ingresos diarios suavizados - Tienda 1')
plt.xlabel('Fecha')
plt.ylabel('Ingresos ($)')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
categoria_tienda1 = tienda['Categoría del Producto'].tolist()
categoria_tienda2 = tienda2['Categoría del Producto'].tolist()
categoria_tienda3 = tienda3['Categoría del Producto'].tolist()
categoria_tienda4 = tienda4['Categoría del Producto'].tolist()

registro_tienda1 = [(categoria,categoria_tienda1.count(categoria)) for categoria in set(categoria_tienda1)]
registro_tienda2 = [(categoria,categoria_tienda2.count(categoria)) for categoria in set(categoria_tienda2)]
registro_tienda3 = [(categoria,categoria_tienda3.count(categoria)) for categoria in set(categoria_tienda3)]
registro_tienda4 = [(categoria,categoria_tienda4.count(categoria)) for categoria in set(categoria_tienda4)]

print(registro_tienda1)
print(registro_tienda2)
print(registro_tienda3)
print(registro_tienda4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

categorias_tiendas = [registro_tienda1,registro_tienda2,registro_tienda3,registro_tienda4]
todas_las_categorias = set(categoria[0]for tienda in categorias_tiendas for categoria in tienda)
cantidad_por_categoria = {categoria:[] for categoria in todas_las_categorias}

for categoria in todas_las_categorias:
  for tienda in categorias_tiendas:
    cantidad = next((c[1] for c in tienda if c[0] == categoria), 0)
    cantidad_por_categoria[categoria].append(cantidad)


ancho_barras = 0.2
posiciones = np.arange(len(todas_las_categorias))

colores = ['b', 'g', 'r', 'c']

fig,ax = plt.subplots(figsize = (12,6))

for i, tienda in enumerate(categorias_tiendas):
  cantidades = [cantidad_por_categoria[categoria][i] for categoria in todas_las_categorias ]
  ax.bar(posiciones + i * ancho_barras, cantidades, width=ancho_barras, label=f'Tienda {i+1}', color=colores[i])


ax.set_xlabel('Categorías')
ax.set_ylabel('Cantidad de productos')
ax.set_title('Cantidad de productos por Categoría y Tienda')


ax.set_xticks(posiciones + ancho_barras * (len(categorias_tiendas) - 1) / 2)
ax.set_xticklabels(todas_las_categorias, rotation=45, ha='right')

ax.legend()

plt.tight_layout()
plt.show()
```
